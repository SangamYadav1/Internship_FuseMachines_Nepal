```python
#Create a pandas series with some random numbers and texts and some null values.

import pandas as pd
import numpy as np

series = pd.Series([1, '5 ' , 'sangam' , 'adarsh' , "null", 3.0] , index = ['A','B','C','D','E','F'])
print(series)
```

```python
#add some data to your series
series_adddata = series.append(pd.Series(data = ['ujjwal',3.0 ] , index = ['G','H']))
print(series_adddata)
```

```python
#change data type to numeric
series_numeric = pd.to_numeric(series_adddata , errors = "coerce")
print(series_numeric)
```

```python
#drop na
dropna_series = series_numeric.dropna()
print(dropna_series)
```

```python
#change data type to integer
series_int = dropna_series.astype(int)
print(series_int)
```

```python
#sort the index of the series alphabetically 
series_int.sort_index(ascending = False)
```

```python
#Count the occurrences of unique items of the series
series_int.value_counts()
```

```python
#Sort the given series in descending order
series_int.sort_values(ascending = False)
```

```python
#Merge both the dataframe using:
#inner
mergedStuff = pd.merge(df1, df2, on=['id'], how='inner')
mergedStuff.head()
```

```python
#outer
mergedStuff = pd.merge(df1, df2, on=['id'], how='outer')
mergedStuff.head()
```

```python
#left
mergedStuff = pd.merge(df1, df2, on=['id'], how='left')
mergedStuff.head()
```

```python
#right
mergedStuff = pd.merge(df1, df2, on=['id'], how='right')
mergedStuff.head()
```

#Differences observed between inner ,outer ,left and right
- The difference observed is:

     - inner will return integer , 
        - outer will return float , 
          - left will return float to left of the data set,
              - ight will return float to right of the data set.

```python
#Read the dataset as pandas dataframe

import pandas as pd
import io
import requests
url= "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/01-01-2021.csv"
s=requests.get(url).content
d=pd.read_csv(io.StringIO(s.decode('utf-8')))
```

```python
d.head(5)
```

```python
#Filter the Country_Region whose value == “India” and assign it in dataframe df_india.
df_india = d[d['Country_Region' ] == 'India']
df_india.head(5)
```

```python
#To print the entire data set
pd.set_option("display.max_rows", None, "display.max_columns", None)
#d
```

```python
#Retrieve the active cases of “Madhya Pradesh” Province from df_india.
df_india.loc[df_india['Combined_Key'] == 'Madhya Pradesh, India', 'Active']
```

```python
##Retrieve the state with the maximum number of confirmed cases from df_india
df_india.Confirmed.max()
```

```python
#Filter the Country_Region whose value == “Canada” and assign it in dataframe df_canada.
df_canada = d[d["Country_Region"] == 'Canada']
df_canada.head(5)
```

```python
#to display the entire data set
pd.set_option("display.max_rows", None, "display.max_columns", None)
#d
```

```python
#Checking the active number of cases in Province/State Diamond Princess
df_canada.loc[df_canada['Combined_Key'] == 'Diamond Princess, Canada' , 'Active']
```

```python
#Province/State Diamond Princess  has the “NaN” value for Active number of cases. Fill it with the median of the Active cases.
df_canada.loc[df_canada['Combined_Key'] == 'Diamond Princess, Canada' , 'Active'] = df_canada.Active.fillna(df_canada.Active.mean())
```

```python
#checking the value after replacing the NaN value with mean
df_canada.loc[df_canada['Combined_Key'] == 'Diamond Princess, Canada', 'Active']
```

```python
#Filter the Country_Region whose value == “US” and assign it in dataframe df_us.
df_us = d[d['Country_Region' ] == 'US']
df_us.head(5)
```

```python
#Groupby the df_us dataframe by ‘State_Province’ and aggregate with maximum values of the attributes and 
#assign it in df_us_states.
df_us_states = df_us.groupby('Province_State').max()
df_us_states
```

```python
# Re-index the dataframe df_us_states.

df_us_states.reset_index(inplace=True)
df_us_states.head()
```

```python
## The United States has 50 states. df_us_states contains additional states which are not the states of the United States.
#Remove those additional states from the dataframe.

states =["Alaska",
                  "Alabama",
                  "Arkansas",
                  "Arizona",
                  "California",
                  "Colorado",
                  "Connecticut",
                  "District of Columbia",
                  "Delaware",
                  "Florida",
                  "Georgia",
                  "Hawaii",
                  "Iowa",
                  "Idaho",
                  "Illinois",
                  "Indiana",
                  "Kansas",
                  "Kentucky",
                  "Louisiana",
                  "Massachusetts",
                  "Maryland",
                  "Maine",
                  "Michigan",
                  "Minnesota",
                  "Missouri",
                  "Mississippi",
                  "Montana",
                  "North Carolina",
                  " North Dakota",
                  "Nebraska",
                  "New Hampshire",
                  "New Jersey",
                  "New Mexico",
                  "Nevada",
                  "New York",
                  "Ohio",
                  "Oklahoma",
                  "Oregon",
                  "Pennsylvania",
                  "Rhode Island",
                  "South Carolina",
                  "South Dakota",
                  "Tennessee",
                  "Texas",
                  "Utah",
                  "Virginia",
                  "Vermont",
                  "Washington",
                  "Wisconsin",
                  "West Virginia",
                  "Wyoming"]
states = pd.Series(states)

states = pd.DataFrame(states)

df_us_states = df_us_states[df_us_states['Province_State'].isin(states[0])]


df_us_states.reset_index(drop = True, inplace = True)
```

```python
# Which state has the minimum number of active cases?
value = df_us_states['Active'].min()
print(value)
df_us_states.loc[df_us_states['Active'] == value , 'Combined_Key']
```
